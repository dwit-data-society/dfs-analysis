```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import streamlit as st
```

```python
rows = []
with open('../data/orders.csv', 'r', encoding='utf-8-sig') as f:
    for line in f:
        # Strip outer quotes
        line = line.strip().strip('"')
        
        # Fix the semicolon inside the item name
        # The pattern is: ""Sadheko Chicken mo;mo""
        line = line.replace('Chicken mo;mo', 'Chicken momo')
        
        # Split on semicolons
        parts = line.split(';')
        # Clean each part - remove surrounding quotes and double-quote escapes
        cleaned_parts = [p.strip('"').replace('""', '"') for p in parts]
        rows.append(cleaned_parts)

# Create DataFrame
orders_data = pd.DataFrame(rows[1:], columns=rows[0])
print(orders_data.head())
print(f"\nShape: {orders_data.shape}")

# Verify the fix
print("\nSample of item names with 'Sadheko':")
sadheko_items = orders_data[orders_data['item_name'].str.contains('Sadheko', na=False)]['item_name'].unique()
print(sadheko_items[:5])
```

```python
orders_data['total_price'] = pd.to_numeric(orders_data['total_price'], errors='coerce')
revenue = orders_data['total_price'].sum()
print(f'The revenue is {revenue}')
```

```python
revenue_per_item = orders_data.groupby('item_name')['total_price'].sum().reset_index()
revenue_per_item.columns = ['item', 'total_revenue']
revenue_per_item
```

```python
N = 10
top_N_revenue = revenue_per_item.sort_values(by='total_revenue', ascending=False).head(N)
```

```python
fig, axes = plt.subplots()

axes.bar(top_N_revenue['item'], top_N_revenue['total_revenue'])
plt.xticks(rotation=90)
```
